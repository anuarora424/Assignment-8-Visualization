5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap?

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set the random seed for reproducibility
np.random.seed(42)

# Generate synthetic dataset with 5 correlated features
n_samples = 100
X1 = np.random.normal(0, 1, n_samples)
X2 = X1 * 0.8 + np.random.normal(0, 0.5, n_samples)  # X2 is correlated with X1
X3 = X1 * 0.5 + X2 * 0.5 + np.random.normal(0, 0.3, n_samples)  # X3 is correlated with X1 and X2
X4 = X2 * 0.6 + np.random.normal(0, 0.7, n_samples)  # X4 is correlated with X2
X5 = X3 * 0.4 + X4 * 0.6 + np.random.normal(0, 0.4, n_samples)  # X5 is correlated with X3 and X4

# Create a DataFrame with these features
data = pd.DataFrame({'X1': X1, 'X2': X2, 'X3': X3, 'X4': X4, 'X5': X5})

# Calculate the correlation matrix
corr_matrix = data.corr()

# Plotting the correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, cbar=True)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

2. Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns?

```python
# Importing necessary libraries
import numpy as np
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
from bokeh.io import output_notebook

# Generating random data
np.random.seed(42)
n = 100
x = np.random.random(n) * 10  # Random x values
y = np.random.random(n) * 10  # Random y values
sizes = np.random.randint(10, 100, n)  # Random sizes for the markers
colors = np.random.choice(['red', 'green', 'blue', 'orange', 'purple'], size=n)  # Random colors

# Creating a ColumnDataSource
source = ColumnDataSource(data={'x': x, 'y': y, 'sizes': sizes, 'colors': colors})

# Creating the scatter plot
output_notebook()
p = figure(title="Random Scatter Plot", x_axis_label='X', y_axis_label='Y', tools="pan,box_zoom,reset")

# Adding the scatter renderer with variable size and color
p.scatter('x', 'y', size='sizes', color='colors', source=source, legend_field='colors', fill_alpha=0.6)

# Show the plot
show(p)
```

```python
To create a heatmap in Bokeh with your dataset, we need to reshape the data and use Bokeh's image function to visualize it. Let's go through each step:

We'll create a Bokeh figure.
Use the image function to add the heatmap.
Customize the color mapper and add labels if needed.
Heres the complete code:
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColorBar, LinearColorMapper
from bokeh.transform import linear_cmap
import numpy as np

# Display Bokeh plots inline
output_notebook()

# Data for the heatmap
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Bokeh figure setup
p = figure(title="Heatmap Example",
           x_axis_label='X', y_axis_label='Y',
           x_range=(0, 1), y_range=(0, 1),
           tools="hover")

# Color mapper for heatmap
color_mapper = LinearColorMapper(palette="Viridis256", low=data_heatmap.min(), high=data_heatmap.max())

# Add the heatmap image
p.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1, color_mapper=color_mapper)

# Adding a color bar
color_bar = ColorBar(color_mapper=color_mapper, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show plot
show(p)
```
